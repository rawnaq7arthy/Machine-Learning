```python
import tensorflow as tf
```

```python
input_layer = tf.keras.layers.Input(shape=[6], name="Input_Layer")
layer_1 = tf.keras.layers.Dense(100, activation='relu', name="Dense_100")(input_layer)
layer_2 = tf.keras.layers.Dense(64, activation='relu', name="Dense_64")(layer_1)
layer_3 = tf.keras.layers.Dense(32, activation='relu', name="Dense_32")(layer_2)
layer_4 = tf.keras.layers.Dense(16, activation='relu', name="Dense_16")(layer_3)
output_layer = tf.keras.layers.Dense(1, name="Output_Layer")(layer_4)

model = tf.keras.models.Model(inputs=input_layer, outputs=output_layer)
```
