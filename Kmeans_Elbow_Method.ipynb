```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
X = np.random.randint(20, 100, size=(30, 2)).astype(float)
print(X.shape)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], color='#FFEB99', edgecolor='black', s=80, alpha=0.7)
plt.title("Random 2D Points", fontsize=14)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
inertias = []

K = range(2, 12)  # Changed from 1–10 to 2–11 for more meaningful clustering

# Generate new random data
X = np.random.randint(10, 200, size=(50, 2)).astype(float)  # More points, wider range

# Fit KMeans with updated settings
for k in K:
    kmeanModel = KMeans(n_clusters=k, n_init=15, max_iter=300, random_state=10)
    kmeanModel.fit(X)
    inertias.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(K, inertias, marker='x', color='slateblue', linestyle='-', linewidth=2, 
         markerfacecolor='red', markeredgecolor='red', alpha=0.9)
plt.title('Elbow Method for Optimal k', fontsize=14)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia (WCSS)', fontsize=12)
plt.xticks(K)
plt.show()
```
